```python
from matplotlib import pyplot as plt
import time
from sendkeys import PressKey,ReleaseKey
import numpy as np
import requests
```

```python
coslistz=[]
coslistx=[]
coslisty=[]
```

```python
def calibrate(x,y,z,a,b,c):
    global coslist
    vec_x=[1,0,0]
    vec_y=[0,1,0]
    vec_z=[0,0,1]
    initial=np.array([x,y,z])
    g_acc=np.array([a,b,c])
    
    total_g=np.subtract(g_acc,initial)
    ct1=(np.dot(vec_z,total_g))/(mag(total_g)*mag(vec_z))
    ct2=(np.dot(vec_x,total_g))/(mag(total_g)*mag(vec_x))
    ct3=(np.dot(vec_y,total_g))/(mag(total_g)*mag(vec_y))
    coslistz.append(ct1)
    coslistx.append(ct2)
    coslisty.append(ct3)
    
    forward(ct1)

    
def mag(a):
    return np.linalg.norm(a)
    
```

```python
datalist=[]
acc1data=[]
acc2data=[]
acc3data=[]

grav_1data=[]
grav_2data=[]
grav_3data=[]

gyr1data=[]
gyr2data=[]
gyr3data=[]
```

```python
def forward(y):
    start=time.time()
    if(abs(y)>.8):
        while((time.time()-start)<1):
            PressKey(0x11)
            time.sleep(.1)
    ReleaseKey(0x11)  
```

```python
def jump(y):
    start=time.time()
    if(y>1):
        while((time.time()-start)<1):
            PressKey(0x39)
            time.sleep(.1)
    ReleaseKey(0x39) 
```

```python
def side(y):
    start=time.time()
    if(y>2):
        while((time.time()-start)<1):
            PressKey(0xCB)
            time.sleep(.1)
    if(y<-2):
        while((time.time()-start)<1):
            PressKey(0xCD)
            time.sleep(.1)
            
    ReleaseKey(0xCB)
    ReleaseKey(0xCD)
```

```python
while(True):

    input_control=requests.get("http://192.168.18.4:8080/")
    input_control=input_control.content.decode("utf-8")
    input_control=input_control.split(" ")
    if(input_control[0]=="closed"):
        break
    
    acc1=float(input_control[0])
    acc1data.append(acc1)
    
    acc2=float(input_control[1])
    acc2data.append(acc2)

    acc3=float(input_control[2])
    acc3data.append(acc3)
  
    grav_1=float(input_control[3])
    grav_1data.append(grav_1)
    
    grav_2=float(input_control[4])
    grav_2data.append(grav_2)

    
    grav_3=float(input_control[5])
    grav_3data.append(grav_3)
    
    
    gyr1=float(input_control[6])
    gyr1data.append(gyr1)
    
    gyr2=float(input_control[7])
    gyr2data.append(gyr2)

    gyr3=float(input_control[8])
    gyr3data.append(gyr3)
    
    side(acc1)
    jump(acc3)
    
    calibrate(acc1,acc2,acc3,grav_1,grav_2,grav_3)
    
#    forward(acc2)
#    side(grav_2)

    list1=[acc1,acc2,acc3,grav_1,grav_2,grav_3,gyr1,gyr2,gyr3]
    datalist.append(list1)
```

```python
fig=plt.figure(figsize=(10,10))
x=range(len(coslistz))

plt.plot(x,coslistz,label="z")
plt.plot(x,coslistx,label="x")
plt.plot(x,coslisty,label="y")

plt.legend(loc="best")
plt.show
```
